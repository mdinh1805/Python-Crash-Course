```python
# USING CONCATENATE FUNCTION TO STACK ARR WITH DIFFERENT SIZE
import numpy as np

# Two arrays of different sizes
array1 = np.array([[1, 2], [3, 4]])
array2 = np.array([[5, 6], [7,8]])

# Concatenate along axis 0 (rows)
result = np.concatenate((array1, array2), axis=0)

print(result)
```

In this case, array1 has a shape of (2, 2) and array2 has a shape of (1, 2). Since both arrays have the same number of columns (axis 1), they can be concatenated along the rows (axis 0).

If you tried to concatenate along axis 1 (columns), you'd get a shape mismatch error because the number of rows would be different.

#**FIGURES AND AXES**

In matplotlib, pyplot is used to create Figures and change the characteristics of figures. The figure contains the overall window where plotting happens, contained within the figure are where actual graphs are plotted. All plotting is done with respect to an Axes (or subplot).


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values and compute y values
x = np.linspace(-10, 9, 29)
y = x**2 - 4*x + 10

# Create a new figure
fig = plt.figure()

# Add axes to the figure
ax = fig.add_axes ([0.1,0.1,0.7,1])

# Plot the function on the axes
plt.plot(x, y)

# Set labels and title
ax.set_xlabel('x')  # x-axis label
ax.set_ylabel('y')  # y-axis label
ax.set_title('Plot of y = x^2 - 4x + 10')

# Show the plot
plt.show()
```

**In the code:**

* The figure method called using pyplot class returns figure object.

* Line 5: We can call add axes method to add an axes to this object.

* fig.add_axes([left, bottom, width, height]): Adds a new set of axes to the figure. The parameters define the position and size of the axes as fractions of the figure's width and height.


The parameters [0.2, 0.2, 0.7, 0.8] passed to the add axes method are the distance from the left and bottom of the default axis (of Figure): [0.2, 0.2]; and the width and height of the axis: 0.7 and 0.8, respectively.

#**CREATING A RANDOM WALK**

```python
from random import choice

# Class to generate a random walk
class RandomWalk:
    def __init__(self, numpoint):
        # Initialize number of points in the walk
        self.numpoint = numpoint

        # All walks start at (0, 0)
        self.x_values = [0]
        self.y_values = [0]

    # Method to fill in the random walk
    def fill_walk(self):
        # Keep taking steps until the walk reaches the desired length
        while len(self.x_values) < self.numpoint:
            # Decide which direction to go and how far
            x_direction = choice([-1, 1])  # Randomly choose left or right
            x_distance = choice([1, 2, 3, 4])  # How far to move in that direction
            x_step = x_direction * x_distance  # Step in the x direction

            y_direction = choice([1, -1])  # Randomly choose up or down
            y_distance = choice([0, 1, 2, 3, 4])  # How far to move in the y direction
            y_step = y_direction * y_distance  # Step in the y direction

            # Reject moves that go nowhere
            if x_step == 0 and y_step == 0:
                continue

            # Calculate the new position
            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step

            # Append the new values to the lists
            self.x_values.append(x)
            self.y_values.append(y)

# Example usage:
numpoint = int(input("Xin moi nhap so: "))
rw = RandomWalk(numpoint)
rw.fill_walk()

print("X values:", rw.x_values)
print("Y values:", rw.y_values)
```

```python
## Plotting
import matplotlib.pyplot as plt
numpoint = int(input('Xin moi nhap so:')) #Number of point in the walk
rw = RandomWalk(numpoint)
rw.fill_walk()
#
plt.style.use('classic')
minh, ax = plt.subplots()
ax.scatter (rw.x_values, rw.y_values, s = 10)
ax.set_aspect('equal')
plt.show()
```
